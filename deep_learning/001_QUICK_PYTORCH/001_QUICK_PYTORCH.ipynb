```python
data3 = torch.FloatTensor([[1., 2., 3.], [4., 5., 6.] ])
print(data3)
print(data3[1, 1])
print(data3[:, :])
print(data3[:1, :2])
# 텐서 인덱싱(슬라이싱)
```

```python
data4 = torch.FloatTensor([ [1, 2, 3], [5, 5, 5] ])
data5 = data4 > 3
print(data5)
print(data4[data5])
print(data4[(data4 > 3) & (data4 < 6)])
# Boolean 인덱싱
```

```python
data1 = torch.randn(4, 3)
print(data1)

print(data1[ [1, 2, 0] ])
print(data1[ [1, -1] ])

print(data1[ [1, 2] ] [:, [0, 2] ])
print(data1[ [1, -1] ] [:, [0] ])
# Fancy 인덱싱(특정 행과 열 추출하기)
```

```python
data1 = torch.FloatTensor([ [1, 2, 3], [4, 5, 6] ])
print(data1)
data2 = data1[:2, :2]
print(data2)

# 얕은 복사
# data2[1, 1] = 4
print(data2)
print(data1)

# 깊은 복사
data2 = data1[:2, :2].clone().detach()
print(data2)
data2[1, 1] = 4
print(data2)
print(data1)
```

```python
import torch

data1 = torch.FloatTensor([7, 2, 0, 4, 1])
index = torch.where(data1 < 3)
print(index)
print(data1[index])
# 조건에 맞는 인덱스 번호 리턴
```

```python
print(data1)
data2 = torch.where(data1 < 3, -1, 1)
print(data2)
# 조건에 맞으면 -1, 틀리면 1로 원소 값을 수정
```

```python
data1 = torch.FloatTensor([ [1, 2, 3], [4, 5, 6] ])
data2 = torch.where(data1 < 3, -1, 1)
print(data2)
# 다차원 배열에도 사용가능
```

```python
data1 = torch.FloatTensor([ [1, 2, 3], [4, 5, 6] ])
print(data1.min())
print(data1.max())
print(data1.sum())
print(data1.mean()) # 평균
print(data1.var()) # 분산
print(data1.std()) # 표준편차
print(data1.argmin()) # 최소값의 인덱스 번호
print(data1.argmax()) # 최대값의 인덱스 번호
```
