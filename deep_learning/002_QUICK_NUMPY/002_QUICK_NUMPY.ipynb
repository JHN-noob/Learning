```python
data1 = np.array([ [1, 2, 3], [4, 5, 6] ])
data2 = np.array([ [1, 2, 3], [4, 5, 6] ])
print(data1+data2)
print(data1-data2)
print(data1*data2)
print(data1/data2)
print(data1%data2)
print(data1//data2)
```

```python
data1 = np.array([ [1, 2], [3, 4], [5, 6] ])
print("data1", data1, data1.shape, "\n")
print(data1 + 1, "\n")
data2 = np.array([ [1], [2], [3] ])
print("data2", data2, data2.shape, "\n")
print(data1 + data2, "\n")
data3 = np.array([ [1, 2, 3] ])
print("data3", data3, data3.shape, "\n")
print(data2 + data3, "\n")
# numpy의 브로드캐스팅
```

```python
data1 = np.random.randn(3, 2)
data2 = np.random.randn(2, 3)
data1 @ data2
# 행렬 곱셈
```

```python
data3 = np.array([ [1., 2., 3.], [4., 5., 6.] ])
print(data3)
print(data3[1, 1])
print(data3[:, 1])
print(data3[:, :])
print(data3[:1, :2])
# 행렬의 인덱싱
```

```python
data1 = np.array([ [1, 2, 3], [4, 5, 6] ])
data2 = data1 > 3
print(data2)
print(data1[data2])
print(data1[(data1 > 3) & (data1 < 5)])
# boolean 인덱싱
```

```python
data1 = np.random.randn(4, 3)
print(data1)

print(data1[ [1, 0, 2] ])
print(data1[ [1, -1] ])

print(data1[ [1, 2] ][:, [0, 2] ])
print(data1[ [1, -1] ][:, [0] ])
# fancy 인덱싱(특정 행과 열 추출하기)
```

```python
data1 = np.array([ [1, 2, 3], [4, 5, 6] ])
print(data1)
data2 = data1[:2, :2]
print(data2)

# data2[1, 1] = 4
print(data2)
print(data1)
# 얉은 복사
```

```python
data3 = data1[:2, :2].copy()
print(data3)
data3[1, 1] = 4
print(data3)
print(data1)
# 깊은 복사
```

```python
data1 = np.array([7, 5, 5, 4, 1])
index = np.where(data1 < 3)
print(index)
print(data1[index])
# 조건에 맞는 값 인덱싱
```

```python
print(data1)
data2 = np.where(data1 < 3, -1, 1)
print(data2)
# 조건에 맞으면 -1, 틀리면 1로 값을 수정
```

```python
data1 = np.array([ [1, 2, 3], [4, 5, 6] ])
data2 = np.where(data1 < 3, -1, 1)
print(data2)
```

```python
data1 = np.array([ [1, 2, 3], [4, 5, 6] ])
print(data1.min())
print(data1.max())
print(data1.sum())
print(data1.mean())
print(data1.var())
print(data1.std())
print(data1.argmin())
print(data1.argmax())
```
